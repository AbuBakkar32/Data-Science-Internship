# Exploratory Data Analysis

```python
# State group wise view numerical data as a mean value
df_state=df.groupby(['State'])[['Quantity','Sales','Discount','Profit']].mean()
df_state
```

```python
# State group wise view numerical data as a min value
df_state=df.groupby(['State'])[['Quantity','Sales','Discount','Profit']].min()
df_state
```

```python
# State group wise view numerical data as a max value
df_state=df.groupby(['State'])[['Quantity','Sales','Discount','Profit']].max()
df_state
```

```python
# State group wise view numerical data as a median value
df_state=df.groupby(['State'])[['Quantity','Sales','Discount','Profit']].median()
df_state
```

```python
sp = df.groupby('State')['Profit'].sum()
```

```python
plt.figure(figsize=(16,5))
sns.barplot(x=sp.index, y=sp.values)
plt.xticks(rotation=90)
plt.ylabel('Profit')
plt.show()
```

```python
# To see the Pair plot
sns.pairplot(df, hue='Region')
```

```python
sns.pairplot(df, hue='Region', diag_kind="hist")
```

```python
sns.pairplot(df,hue='Category')
```

```python
sns.pairplot(df,hue='Sub-Category', diag_kind="hist")
```

```python
sns.countplot("Ship Mode",data=df)
```

```python
sns.countplot("Segment",data=df)
```

```python
sns.countplot("Region",data=df)
```

```python
sns.countplot("Category",data=df)
```

```python
# To analysing the total number of orders from different states
state = df['State'].value_counts()
plt.figure(figsize=(15,16))
sns.countplot(x=df['State'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(5,5))
df.groupby("Category")["Profit"].sum().plot.pie(autopct="%1.1f%%", shadow=True)
plt.title('Category wise Profit analysis',fontsize=20)
```

```python
# To see the Region wise different sub-categories order
plt.figure(figsize=(10,6))
sns.countplot(x="Sub-Category", hue="Region",data=df)
plt.xticks(rotation="vertical")
plt.show()
```

```python
df_order = df.groupby(['State','City']).size().reset_index(name='total_orders')
```

```python
df_order[df_order['total_orders']==df_order['total_orders'].min()]
```

```python
plt.figure(figsize=(20,22))
df.groupby("State")["Sales"].sum().plot.pie(autopct="%1.1f%%")
plt.title('State wise of Sales Analysis',fontsize=20)
plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize = (16,7))
sns.distplot(df['Sales'], ax = axs[0,0])
sns.distplot(df['Quantity'], ax = axs[0,1])
sns.distplot(df['Discount'], ax = axs[1,0])
sns.distplot(df['Profit'], ax = axs[1,1])
```

```python
pd.crosstab( df['Region'],df['Category']).plot(kind='bar', figsize=(18,5))
plt.xticks(rotation=45)
plt.plot()
```

```python
sales_of_subCategory = df.groupby('Sub-Category')['Sales'].sum()
plt.figure(figsize=(12,4))
sns.barplot(x=sales_of_subCategory.index, y=sales_of_subCategory.values)
plt.xticks(rotation=90)
plt.show()
```
